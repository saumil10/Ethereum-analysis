```python
import pandas as pd
```

```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
df1 = pd.read_csv('/home/jovyan/teaching_material/ECS765P/cousework/PART A/avg_txn.csv')
```

```python
df1.columns = ['Date', 'Average Transactions each month']
```

```python
df1.head()
```

```python
# remove brackets and comma from the 'Date' column
df1['Date'] = df1['Date'].str.replace('[^\w\s]', '').str.strip()

# convert the 'Date' column to datetime format
df1['Date'] = pd.to_datetime(df1['Date'], format='%m%Y')

# print the resulting DataFrame
print(df1)
```

```python
plt.figure(figsize = (20,10))
plt.xticks(rotation=90)
plt.xlabel('Month/Year')
plt.ylabel('Average Transactions each month')
x=df1['Date']
y=df1['Average Transactions each month']
plt.bar(x,y,color='orange',width=30)
```
